```python
# Function to convert latitude values
def convert_latitude(lat):
    if isinstance(lat, str):  # Check if the value is a string
        if 'S' in lat:  # Southern hemisphere, make it negative
            return -float(lat.replace('S', ''))
        elif 'N' in lat:  # Northern hemisphere, keep it positive
            return float(lat.replace('N', ''))
        else:
            return float(lat)  # If there is no direction
    return lat  # If it's already a float, return it as is

# Function to convert longitude values
def convert_longitude(lon):
    if isinstance(lon, str):  # Check if the value is a string
        if 'W' in lon:  # Western hemisphere, make it negative
            return -float(lon.replace('W', ''))
        elif 'E' in lon:  # Eastern hemisphere, keep it positive
            return float(lon.replace('E', ''))
        else:
            return float(lon)  # If there is no direction
    return lon  # If it's already a float, return it as is

# Apply conversion functions to the Latitude and Longitude columns
df['Latitude'] = df['Latitude'].apply(convert_latitude)
df['Longitude'] = df['Longitude'].apply(convert_longitude)

# Check the data types after conversion
print(df['Latitude'].dtype)  # Should be float
print(df['Longitude'].dtype)  # Should be float
```

```python
from scipy.stats import ttest_ind

# Split data into two periods: before 1950 and after 1950
before_1950 = df[df['Date'].dt.year < 1950]['AverageTemperature']
after_1950 = df[df['Date'].dt.year >= 1950]['AverageTemperature']

# Perform t-test
t_stat, p_val = ttest_ind(before_1950, after_1950, nan_policy='omit')
print(f"T-statistic: {t_stat}, P-value: {p_val}")
```

```python
# Compute rolling mean for temperature with a 10-year window
df['RollingMean_Temp'] = df['AverageTemperature'].rolling(window=120, min_periods=1).mean()

# Plot rolling mean temperature
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['RollingMean_Temp'], label='10-Year Rolling Average', color='orange')
plt.title('10-Year Rolling Average of Global Temperature')
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Preparing data for linear regression
X = df[['Latitude']]  # Independent variable
y = df['AverageTemperature']  # Dependent variable

# Fit the linear regression model
reg = LinearRegression()
reg.fit(X, y)

# Get the slope and intercept
slope = reg.coef_[0]
intercept = reg.intercept_
print(f"Slope: {slope}, Intercept: {intercept}")

# Plot regression line
plt.figure(figsize=(10, 6))
plt.scatter(df['Latitude'], df['AverageTemperature'], alpha=0.5, label='Data Points')
plt.plot(df['Latitude'], reg.predict(X), color='red', label='Regression Line')
plt.title('Latitude vs Average Temperature (with Regression Line)')
plt.xlabel('Latitude')
plt.ylabel('Average Temperature (°C)')
plt.legend()
plt.grid(True)
plt.show()
```
